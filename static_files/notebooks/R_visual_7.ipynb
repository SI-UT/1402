<font color=skyblue size=4 id="t1"> <b>Introduction to ggplot2</b></font>
<hr />
<br>

<p align='justify'><b>ggplot2</b> is a system for declaratively creating graphics, based on The Grammar of Graphics. The grammar of graphics is a way of thinking about how graphs are constructed that allows data analysts to move beyond thinking about a small number of graph types. <p>

<p align='justify'>There are seven structures or layers in the grammar of graphics. We can layer on just a minimum number of structures or make more complicated graphs by using multiple types of layers and multiple layers of some of the types. Th seven layers are as followed:<p>

1. <p align='justify'><b>Data:</b> A data frame with one or more variables, each with one or more observations.</p>

2. <p align='justify'><b>Aesthetic:</b> A mapping of one or more variables to one or more visual elements on the graph. For example, mapping variables to x-axis, y-axis and color. </p>

3. <p align='justify'><b>Scale:</b> A mapping to scale the potential values, use a specific scale to represent multiple values or a range.</p>

4. <p align='justify'><b>Geometry:</b> The type or shape of the visual elements on the graph. For example, a point in the case of a scatter plot or a bar in a bar plot.</p>

5. <p align='justify'><b>Statistics:</b> Showing statistical measures in the visualization like measures of central tendency, spread, confidence intervals.</p>

6. <p align='justify'><b>Facets:</b> Creating subplots based on specific data dimensions. For example, plotting the distribution of some variables in different groups of another variable .</p>

<img width=300 align=center src="https://www.datanovia.com/en/wp-content/uploads/2020/05/062-add-p-values-to-ggplot-facets-multipanel-box-plots-facet-wrap-1.png" style="margin:0;" />


7. <p align='justify'><b>Coordinate System:</b> The kind of coordinate system the visualization should be based on. For example, polar, fixed and flipped.</p>

<p align='justify'>* First, second and fourth elements are neccessary for all plots. Other layers can be added if needed or preffered.</p>

### <b> Bar plot</b>

```r
ggplot(diamonds, aes(color)) + geom_bar()
```

```r
# adding axis labels and title to the plot
ggplot() + geom_bar(data=diamonds, aes(color)) + xlab("Diamond Color") + 
           labs(y="Count", title="Diamond color Count")
```

```r
# color bars and change text size for better readability, move title to the center
ggplot() + geom_bar(data=diamonds, aes(color, fill=color)) + xlab("Diamond color") + 
           labs(y="Count", title="Diamond color Count") + theme(text = element_text(size = 20), 
                                                                plot.title = element_text(hjust = 0.5)) 
```

```r
base_bar <- ggplot(data=diamonds, aes(color, fill=color)) + geom_bar(width=0.8, color="white") + 
            xlab("Diamond color") + labs(y="Count", title="Diamond color Count") + 
            theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5))

base_bar + coord_flip()
```

```r
# add count to the top of each bar + removing legend
base_bar + geom_text(stat='count', aes(label=..count..), vjust=-1, color="blue", size=3.5) + 
           theme(legend.position="none")
```

### <b>Pie chart</b>

```r
data <- data.frame(
  group=LETTERS[1:5],
  value=c(13,7,9,21,2)
)

# Basic piechart
ggplot(data, aes(x="", y=value, fill=group)) +
  geom_bar(stat="identity", width=1, color="white") +
  coord_polar("y", start=0) 
```

### <b> Box plot</b>

```r
ggplot(diamonds, aes(x=price)) + geom_boxplot() + coord_flip()
```

```r
ggplot(diamonds, aes(x=price)) + geom_boxplot(outlier.colour="red",outlier.size=0.3) + 
                                 coord_flip()
```

```r
ggplot(diamonds, aes(x=cut, y=price, col=cut)) + geom_boxplot(outlier.colour="red",outlier.size=0.3) +
      labs(x="Cut", y="Price", title="Price Distribution over Cut") + 
      theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5), 
            axis.text.x = element_text(angle = 45)) #rotating xticks
```

```r
ggplot(diamonds, aes(x=cut, y=price, fill=color)) + geom_boxplot(outlier.colour="red",outlier.size=0.3) +
      labs(x="Cut", y="Price", title="Price Distribution over Cut") + 
      theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5), 
            axis.text.x = element_text(angle = 45)) #rotating xticks
```

```r
# adding facet and clearing xticks
ggplot(diamonds, aes(x=cut, y=price, fill=cut)) + geom_boxplot() +
      labs(x="Cut", y="Price", title="Price Distribution over Cut") + 
      theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5), 
            axis.text.x = element_blank()) + 
      facet_wrap(~color)
```

### <b> Violin plot</b>

```r
ggplot(diamonds, aes(x=cut, y=price, fill=cut)) + geom_violin() +
      labs(x="Cut", y="Price", title="Price Distribution over Cut") + 
      theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5), 
            axis.text.x = element_text(angle = 45), legend.position="none")
```

```r
ggplot(diamonds, aes(x=cut, y=price, fill=cut)) + geom_violin() +
       geom_boxplot(width=0.1, color="grey", alpha=0.2) +
       labs(x="Cut", y="Price", title="Price Distribution over Cut") + 
       theme(text = element_text(size = 20), plot.title = element_text(hjust = 0.5), 
            axis.text.x = element_text(angle = 45), legend.position="none")
```

### <b> Density Plot</b>

```r
library(ggplot2)
ggplot(diamonds, aes(x=price)) + 
  geom_density(fill="#69b3a2", color="blue") + 
  geom_vline(aes(xintercept=mean(price)),
            color="blue", linetype="dashed", linewidth=1) + 
  theme(text = element_text(size = 20))
```

```r
ggplot(data=diamonds, aes(x=price, group=cut, fill=cut)) +
  geom_density() + theme_minimal() + theme_classic() + 
  theme(text = element_text(size = 20))
```

```r
ggplot(data=diamonds, aes(x=price, group=cut, fill=cut)) +
  geom_density(alpha=0.3) + theme_minimal() + theme_classic() + 
  theme(text = element_text(size = 20))
```

```r
# Using Small multiple
ggplot(data=diamonds, aes(x=price, group=cut, fill=cut)) +
    geom_density() +
    facet_wrap(~cut) +
    theme(legend.position="none", text = element_text(size = 20),
          axis.text.x = element_blank())
```

### <b> Histogram </b>

```r
library(ggplot2)
head(diamonds)
```

```r
ggplot(diamonds, aes(x=depth)) + geom_histogram() + 
  theme(text = element_text(size = 20))
```

```r
ggplot(diamonds, aes(x=depth))+
  geom_histogram(binwidth=2, color="darkblue", fill="lightblue") + 
  geom_vline(aes(xintercept=mean(depth)), color="blue", linewidth=1) + 
  theme(text = element_text(size = 20))
```

```r
ggplot(diamonds, aes(x=depth, fill=cut))+
  geom_histogram(binwidth=2, alpha=0.2) + 
  theme(text = element_text(size = 20))
```

```r
ggplot(diamonds, aes(x=price, fill=cut)) + 
  geom_histogram(aes(y=..density..),alpha = 0.4) + #..density..
  geom_density(alpha = 0.1) + # erase
  theme(text = element_text(size = 20))
```
