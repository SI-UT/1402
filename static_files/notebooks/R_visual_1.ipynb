<b>Qualitative</b> data is used to represent some characteristics or attributes of the data.
<ul><li align='justify'>Nominal Data: label the variables without providing the numerical value for them. Nominal data attributes can’t either be ordered or measured. Examples of nominal data is gender, religion, city and etc.</li>
<li align='justify'>Ordinal Data: the type of data that follows a natural order. Examples of ordinal data include grades of A, B and C, economic state of low, medium and high and etc.</li></ul>

<p align='justify'>As mentioned previously, qualitative data can not be represented itself. Therefore, statistical properties of the data is visualized. Plots that can be used to visualize qualitative data are: contingency table, pie chart, bar plot and box plot.</p>

<b>Quantitative</b> data can be measured and is not just observable. The measurement of data is numerically recorded and represented.
<ul><li align='justify'>Discrete Data: can only attain certain specific values. Examples include number of rooms in a house, number of students in class and etc.</li>
<li align='justify'>Continuous Data: can contain values between a certain range that is within the highest and lowest values. Examples of continuous data are height and weight of students, price of a house and room temperature.</li></ul>

<p align='justify'>As quantitative data consists of numbers, plots used to visualize it include: histograms, line plot, dot plot and scatter plot.</p>
Some of these plots are also used to show the relation between different data with different types.

<font color=skyblue size=4 id="t2"> <b>Contingency Table</b></font>
<hr />
<br>

<p align='justify'>One of the easiest ways to visualize data, specially categorical data, is to use contingency table. The table is very easy to understand for every one and can be used where audience is not completely of one type. The function used to plot a table in R is:</p>
<p align='center'><font size=5><code>table(X)</code></font></p>
where X is a vector of categorical data. 

```r
# producing sample data
set.seed(20)
X <- sample(c("Yes", "No"), 10, replace = TRUE)
# "No" "Yes" "Yes" "No"  "No" "Yes" "No" "Yes" "No" "No"
Y <- sample(c("Europe", "America", "Africa"), 10, replace = TRUE)
# "Europe" "Africa"  "Africa"  "Europe"  "Africa" 
# "Europe" "Europe"  "Europe"  "America" "America"

table_x <- table(X)
table_y <- table(Y)
table_xy <- table(X, Y)

table_x 
table_y 
table_xy
```

```r
# changing column and row names of the table
rownames(table_xy) <- c('N', 'Y')
colnames(table_xy) <- c('Af', 'Am', 'Eu')
table_xy
```

```r
data.frame(table_xy)
```

```r
# set sum of each row to one with adding margin=1
prop.table(table_xy, margin=1)
```

```r
addmargins((prop.table(table_xy, margin=1) * 100))
```

```r
addmargins(round(prop.table(table_xy, margin=1) * 100))
```

<font color=skyblue size=4 id="t3"> <b>Pie Chart</b></font>
<hr />
<br>

<p align='justify'>One of the most used plots to visualize the relative frequenvy of different categories of quantitave data is pie chart. This plot is used widely and therefore most people are familiar with it. The fact that pie chart shows the relative frequency of a category with respect to the whole or other categories is both the strength and weakness of this plot. If there are only a few categories in the data and there is a noticable difference in the frequency of different categories, pie chart is OK. However, in case of high number of categories and very close percentages, pie chart would not be easy to read. To plot pie chart in simple R, the below code can be used: </p>
<p align='center'><font size=5><code>pie(x, labels, radius, main, col, clockwise)</code></font></p>

<b>x:</b> vector of numeric values used in the pie chart

<b>labels:</b> description of the slices

<b>radius:</b> the radius of the circle of the pie chart.(value between −1 and +1)

<b>main:</b> title of the chart

<b>col:</b> color palette

<b>clockwise:</b> a logical value indicating if the slices are drawn clockwise or anti clockwise

```r
# Create data for the graph.
x <- c(21, 62, 10, 53)
labels <- c("London", "New York", "Singapore", "Mumbai")
pie(x, labels, radius=0.8, main="City pie chart")
```

```r
library(ggplot2)
data('diamonds')
head(diamonds)
```

```r
diamond_colors = table(diamonds$color)
colors = as.vector(names(diamond_colors))
pie(diamond_colors, main="Color frequency of diamonds", col=rainbow(length(colors)), labels=colors)
```

```r
# adding frequencies to the pie chart
# color frequencies and names
diamond_colors = table(diamonds$color)
# color names
colors = as.vector(names(diamond_colors))
# color frequncies
col_freq = as.vector(as.numeric(diamond_colors))
# color percentages
slices_labels <- round(col_freq/sum(col_freq) * 100, 1)
# setting slice lables as the frequncy percentage
slices_labels <- paste(slices_labels, "%", sep="")
pie(col_freq, main="Color frequency", col=rainbow(length(colors)), labels=slices_labels, cex=1.2) #cex for font size
```

```r
pie(col_freq, main="Color frequency", col=rainbow(length(colors)), labels=slices_labels, cex=1.2)
# adding legend
legend(-1.2,.5, colors, fill = rainbow(7))
```
