<font color=skyblue size=4 id="t4"> <b>Bar Plot</b></font>
<hr />
<br>

<p align='justify'>As mentioned previously, pie charts are a little hard to read and interpret in some cases and they are not very formal looking. The plot used to overcome the mentioned problems is bar plot. Bar plots are easy to understand and can be used with any type of audience. The function used to plot bar plots in simple R is:</p>
<p align='center'><font size=5><code>barplot(H,xlab,ylab,main, names.arg,col)</code></font></p>

<b>H:</b> a vector or matrix containing numeric values used in bar chart

<b>xlab:</b> label for x axis

<b>ylab:</b> label for y axis

<b>main:</b> title of the bar chart

<b>names.arg:</b> vector of names appearing under each bar

<b>col:</b> give colors to the bars in the graph


<font color=skyblue size=3 id="t4.1"> <b>Simple Vertical Bar Plot</b></font>


```r
barplot(x, xlab="cities", ylab="count", main="City Bar Plot", names.arg=labels)
```

```r
barplot(diamond_colors, xlab="color", ylab="freq", main="Diamond Color Frequencies", 
        names.arg=names(diamond_colors), col=rainbow(length(colors)),
        border="red", density=c(5,10,20,30,7) , angle=c(0,45,90,11,36),
        width=c(0.5,0.2,2,1.5,0.3) )
```


<font color=skyblue size=3 id="t4.2"> <b>Simple Horizontal Bar Plot</b></font>


```r
# horiz=TRUE to make plot horizontal
# cex.names=1.5 to control labels font size
barplot(diamond_colors, xlab="color", ylab="freq", main="Diamond Color Frequencies", names.arg=names(diamond_colors), 
        col=rainbow(length(colors)), horiz=TRUE, cex.names=1.5, cex.main=1.5)
```


<font color=skyblue size=3 id="t4.3"> <b>Vertical Grouped Bar Plot</b></font>


```r
# Grouped Bar Plot with beside=TRUE
counts <- table(diamonds$clarity, diamonds$color)
barplot(counts, main="Diamond Distribution by Color and Clarity",
  xlab="color", ylab='count', col=rainbow(length(unique(diamonds$clarity))),
  legend = rownames(counts), beside=TRUE) #, horiz=TRUE
```

```r
# Stacked Bar Plot
barplot(counts, main="Diamond Distribution by Color and Clarity",
  xlab="color", ylab='count', col=rainbow(length(unique(diamonds$clarity))),
  legend = rownames(counts))
```

<font color=skyblue size=4 id="t5"> <b>Box Plot</b></font>
<hr />
<br>

<p align='justify'>Boxplots are a measure of how well distributed is the data in a data set. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.</p>
<p align="center">
  <img width=600 src="https://media.geeksforgeeks.org/wp-content/uploads/20201127012952/boxplot.png" />
</p>
<p align='justify'>Boxplots can be plotted using the below function:</p>

<p align='center'><font size=5><code>boxplot(x, data, notch, varwidth, names, main)</code></font></p>

<b>x:</b> avector or a formula

<b>data:</b> the dataframe

<b>notch:</b> a logical value, set as TRUE to draw a notch

<b>varwidth:</b> logical value to draw width of the box proportionate to the sample size

<b>names:</b> vector of names appearing under each box

<b>main:</b> give title to graph

```r
boxplot(diamonds$price, main="Diamond Price Distribution")
```

```r
boxplot(diamonds$price,
main = "Diamods Pricce Distribution",
xlab = "price",
ylab = "diamond",
col = "orange",
border = "brown",
horizontal = TRUE,
notch = TRUE
)
```

```r
boxplot(diamonds$price ~ diamonds$color,
main = "Diamods Pricce Distribution",
xlab = "price", ylab = "diamond",
col = rainbow(length(unique(diamonds$clarity))),
notch = TRUE, varwidth=TRUE)
```
