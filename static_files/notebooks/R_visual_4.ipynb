<b>Qualitative</b> data is used to represent some characteristics or attributes of the data.
<ul><li align='justify'>Nominal Data: label the variables without providing the numerical value for them. Nominal data attributes canâ€™t either be ordered or measured. Examples of nominal data is gender, religion, city and etc.</li>
<li align='justify'>Ordinal Data: the type of data that follows a natural order. Examples of ordinal data include grades of A, B and C, economic state of low, medium and high and etc.</li></ul>

<p align='justify'>As mentioned previously, qualitative data can not be represented itself. Therefore, statistical properties of the data is visualized. Plots that can be used to visualize qualitative data are: contingency table, pie chart, bar plot and box plot.</p>

<b>Quantitative</b> data can be measured and is not just observable. The measurement of data is numerically recorded and represented.
<ul><li align='justify'>Discrete Data: can only attain certain specific values. Examples include number of rooms in a house, number of students in class and etc.</li>
<li align='justify'>Continuous Data: can contain values between a certain range that is within the highest and lowest values. Examples of continuous data are height and weight of students, price of a house and room temperature.</li></ul>

<p align='justify'>As quantitative data consists of numbers, plots used to visualize it include: histograms, line plot, dot plot and scatter plot.</p>
Some of these plots are also used to show the relation between different data with different types.

<p align='justify'>Strip charts in R are visualizations or plots that illustrate the distribution of a single continuous variable, usually measured on one axis. A plot is created by placing points along the axis for each data point, with the height or position of the points indicating the variable's value. By using different colors or symbols for each group, strip charts can compare the distributions of the same variable across different groups or categories.</p>

Let us use the ggplot dataset `diamonds`, which contains the prices and other attributes of almost 54,000 diamonds, and the built-in dataset `airquality` which has â€œDaily air quality measurements in New York, May to September 1973.â€-R documentation.

### Chart methods

```r
set.seed(1)
x <- round(runif(100, 0 , 10))

stripchart(x, pch = 19, method = "stack",
           col = 4, main = "method = 'stack'")
```

```r
stripchart(airquality$Ozone, 
           method="jitter",
           main = "method = 'jitter'",
           col="orange",
           jitter = 0.01,
           pch=1)
```

### Multiple Strip Charts

To compare, generate a normal distribution with the same mean and standard deviation of Ozone.

```r
Ozone <- airquality$Ozone
OzoneNorm <- rnorm(200,mean=mean(Ozone, na.rm=TRUE), sd = sd(Ozone, na.rm=TRUE))

x <- list("Ozone"=Ozone, "norm"=OzoneNorm)
```

```r
stripchart(x,
main="Multiple stripchart for comparision",
ylab="Ozone",
method="jitter",
col=c("blue","red"),
pch=16,
las = 2
)
```

```r
?airquality
```

### Strip Chart from Formula

```r
stripchart(Ozone ~ Month,
    main = "Different strip chart for each month",
    group.names=c("May","June","July","August","September"),
    vertical = TRUE, 
    log = "y",
    col = grey.colors(5) ,
    method = "jitter",
    data = airquality)
```

<font color=skyblue size=4 id="t2"> <b>Dot Plots</b></font>
<hr />
<br>

In R, a dot plot displays the distribution of a continuous (numeric) variable. Each observation is represented by a dot in a row or column, with the number of dots representing the frequency. The dot plot is an effective way to visualize the location and spread of a dataset, including outliers and gaps. In R, dot plots can be created using the `dotchart` function.

Creation sample data

```r
day <- c( "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
hours <- c(7, 9, 8, 6, 5, 7, 3)
paper <- c(640, 920, 760, 610, 440, 695, 208)
term <- c(rep(1, 2), rep(2, 3), rep(3, 2))

data <- data.frame(day, hours, paper, term)
data
```

```r
dotchart(data$paper, labels = data$day, pch = 21, bg = "green", pt.cex = 1.5)
```

### Dot plot by group

```r
colors <- numeric(3)

colors[term == "1"] <- "red"
colors[term == "2"] <- "blue"
colors[term == "3"] <- "green"

dotchart(data$hours, labels = data$day, pch = 19,
         pt.cex = 1.5, groups = data$term, color = colors)
```

<font color=skyblue size=4 id="t2"> <b>Scatter Plots</b></font>
<hr />
<br>

A scatter plot in R is a graphical representation of the relationship between two quantitative variables. It depicts data points as markers on a two-dimensional plane, with one variable on the x-axis and the other on the y-axis. Scatter plots are useful for identifying patterns or trends in data, exploring the strength and direction of relationships between variables, and identifying outliers or influential points that may affect the analysis. In R, scatter plots can be created using the `plot()` function.

Creation sample data

```r
options(repr.plot.width=8, repr.plot.height=6)
```

```r
set.seed(1)

n <- 300

x <- runif(n) * 4
eps <- rnorm(n, 0, 3)

y <- exp(x) + eps
```

```r
plot(y ~ x, pch = 19, col = "black", main = "Multivariate Scatter Plot")
```

### Scatter plot with regression line

```r
plot(x, y, pch = 19, col = "gray52")

abline(lm(y ~ x), col = "red", lwd = 4, lty = 2)
lines(lowess(x, y), col = "blue", lwd = 4)


legend("topleft", 
       legend = c("Linear", "Smooth"),
       lwd = 3, lty = c(2, 1), 
       col = c("red", "blue"))
```

### Add correlation

Create the plot and add the calculated value

```r
Corr <- cor(x, y)

plot(x, y, pch = 19)
text(paste("Correlation:", round(Corr, 2)), x = .5, y = 50)
```

### Add multiple series

```r
plot(x, y, pch = 19)

eps <- rnorm(n, 0, .9)
y2 <- x**2 + eps

points(x, y2, col = "orange", pch = 19)
```

<font color=skyblue size=4 id="t2"> <b>Histogram</b></font>
<hr />
<br>

In R, a histogram represents the distribution of numerical data. Bar charts show how data points are distributed across different values. The x-axis of a histogram represents the range of measured or observed values, while the y-axis represents the frequency or proportion of data points within each bin. Using R's `hist()` function, users can create histograms by adjusting parameters like bin width, borders, and colors. Analyzing histograms is useful for exploring the shape and distribution of data, identifying outliers, and assessing overall patterns of variation.

```r
h <- hist(diamonds$price,main="Distribution of price",xlab="price", col="darkmagenta")
h
```

### Use Histogram return values for labels using text()

We see that an object of class histogram is returned which has:
- breaks: places where the breaks occur.
- counts: the number of observations falling in that cell.
- density: the density of cells.
- mids: the midpoints of cells.
- xname: the x argument name.
- equidist: a logical value indicating if the breaks are equally spaced or not.

For example, in the following example we use the return values to place the counts on top of each cell using the text() function.

```r
h <- hist(diamonds$price,main="Distribution of price",xlab="price" ) # breaks: actual number of cells plotted
text(h$mids,(h$counts),labels=h$counts, adj=c(0.5, -0.5)) # x, y 
```

We can also define breakpoints between the cells as a vector. This makes it possible to plot a histogram with unequal intervals. In such case, the area of the cell is proportional to the number of observations falling inside that cell.

```r
hist(diamonds$price,
     main="Distribution of price",
     col="chocolate",
     border="brown",
     xlim=c(0,19000),
     breaks=c(0, 900, 5000, 6000, 9000, 13000, 14000, 15000, 18000, 19000))
```

<font color=skyblue size=4 id="t2"> <b>Density plot</b></font>
<hr />
<br>

A density plot in R is a graphical representation of the distribution of a dataset. It is a smoothed version of a histogram and can be used to visualize the shape of the distribution of a continuous variable. The x-axis represents the values of the variable, and the y-axis represents the density of the distribution. It is created using the `density()` function in R.

```r
x <- rnorm(500)
```

```r
par(mfrow = c(1, 2))

hist(x, main = "Histogram and density", prob = TRUE)

lines(density(x), lwd = 2, col = "red")

plot(density(x), lwd = 2, col = "red", main = "Density")
```

A kernel density estimate is plotted when you plot a probability density function in R. Kernel density plots are non-parametric and require a bandwidth. The density function allows you to set the bandwidth using the `bw` argument.

The kernel density estimation in R is over-smoothed (overfit) when bandwidth is large, and under-smoothed (underfit) when bandwidth is small.

```r
par(mfrow = c(1, 2))

# Large bandwidth
plot(density(x, bw = 20), lwd = 2,
     col = "red", main = "Too big bandwidth")

# Small bandwidth
plot(density(x, bw = 0.05), lwd = 2,
     col = "red", main = "Too small bandwidth")
```

In base R you can use the polygon function to fill the area under the density curve. If you use the `rgb` function in the col argument instead using a normal color, you can set the transparency of the area of the density plot with the `alpha` argument, that goes from 0 to all transparency to 1, for a total opaque color.

```r
set.seed(22)
y <- rnorm(500) + 2
```

```r
plot(density(x), lwd = 2, col = "red", xlim = c(-5, 10), ylim = c(0, 0.5), main = "", xlab = "", ylab="density")
polygon(density(x), col = "red")

lines(density(y), lwd = 2, col = "white")
polygon(density(y), col = rgb(0, 0, 1, alpha = 0.5))
```

<p align='justify'>Boxplots are a measure of how well distributed is the data in a data set. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.</p>

<p align='justify'>Boxplots can be plotted using the below function:</p>

<p align='center'><font size=5><code>boxplot(x, data, notch, varwidth, names, main)</code></font></p>

<b>x:</b> avector or a formula

<b>data:</b> the dataframe

<b>notch:</b> a logical value, set as TRUE to draw a notch

<b>varwidth:</b> logical value to draw width of the box proportionate to the sample size

<b>names:</b> vector of names appearing under each box

<b>main:</b> give title to graph

```r
options(repr.plot.width=6, repr.plot.height=5)
```

```r
boxplot(diamonds$price, main="Diamond Price Distribution")
```

```r
boxplot(diamonds$price,
        main = "Diamods Pricce Distribution",
        xlab = "price",
        ylab = "diamond",
        col = "orange",
        border = "brown",
        horizontal = TRUE,
        notch = TRUE)
```

If your dataset has a categorical variable containing groups, you can create a boxplot from formula. In this example, we are going to use the base R chickwts dataset.

```r
boxplot(diamonds$price ~ diamonds$color,
        main = "Diamods Pricce Distribution",
        xlab = "price", 
        ylab = "diamond",
        col = rainbow(length(unique(diamonds$clarity))),
        notch = TRUE,
        varwidth=TRUE)
```
